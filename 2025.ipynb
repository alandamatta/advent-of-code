# Day 1 - Part 1

```python
pos = 50
rst = 0
with open('input-1.txt', 'r') as f:
    for line in f:
        direction=str(line[:1])
        steps=int(line[1:])
        if direction is "R":
            pos = pos + steps
        else:
            pos = pos - steps
        pos = pos % 100
        if pos == 0:
            rst = rst + 1

print(rst)
```
